# Extract (Extração)

```python
#como a API que eu peguei, não tem modo GET por ID, eu tive que trocar por username


import pandas as pd
df = pd.read_csv('/content/SDW.csv')
user_username = df['Username'].tolist()
print(user_username)
```

```python
import requests
import json

#'GET' API endpoint
swd_api_url = "https://petstore.swagger.io/v2"

#criando uma função pra recuperar os dados na api
def get_user(username):
  response = requests.get(f'{swd_api_url}/user/{username}')
  return response.json() if response.status_code == 200 else "deu ruim"

# percorrendo a lista de users

users = [users for username in user_username if (users := get_user(username)) is not None]
print(json.dumps(users, indent=1))
```

# Transform (transformação)

```python
import random

def generate_ai_news(user):
  try:
    # abre o arquivo e ler as linhas
    with open('/content/mensagens.txt', 'r', encoding='utf-8') as f:
      mensagens = f.readlines()

    # escolhe uma mensagem aleatória e remove espaços extras(strip) da lista
    mensagem_aleatoria = random.choice(mensagens).strip()

    #personaliza com o nome do usuario
    return f'{mensagem_aleatoria}'
  except FileNotFoundError:
    return f'Erro: O arquivo mensagens.txt não foi encontrado.'

# percorrendo os users para gerar a mensagem e adicionar em news
for user in users:
  news = generate_ai_news(user)


  user['firstName'] = news
  print(f" {user['username']}: {news}")
```

# load (carregamento)

```python
# atualizando a lista news de cada usuario na API com a nova mensagem gerada

# --- LOAD ---
def update_user(user):
    url = f"https://petstore.swagger.io/v2/user/{user['username']}"
    response = requests.put(url, json=user)
    return response.status_code

# como a api do swagger não tenho permissão pra criar uma nova chave, eu adicionei a mensagem ao firstname
for user in users:
    status = update_user(user)
    if status == 200:
        print(f"Sucesso! Usuário {user['username']} atualizado.")
    else:
        print(f"Erro {status}: Usuário {user['username']} não encontrado ou erro na API.")
```
